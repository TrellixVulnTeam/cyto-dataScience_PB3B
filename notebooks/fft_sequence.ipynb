```python
%matplotlib widget
im = cv2.imread('RK1.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2YUV)
y, u, v = cv2.split(im)
blur = cv2.GaussianBlur(y,(51,51),0)
trans = cv2.transpose(blur)

edges = cv2.Canny(trans,2,6)
df = pd.DataFrame(data=edges)
df.to_csv('file_name.csv')
plt.figure('y')
plt.imshow(trans, cmap='gray')
plt.figure('Canny')
plt.imshow(edges, cmap='gray')
plt.show()
df
```

```python
%matplotlib widget
gen_im = img.GenImage(16,16, np.uint8)
gen_im.bw_checkerboard(2)
gen_im.show_im()

f = np.fft.fft2(gen_im.image)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))

print(f)

plt.figure("FFT")
plt.imshow(magnitude_spectrum, cmap='gray', vmin=0,vmax=255)
plt.show()
```
